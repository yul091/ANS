```python
def get_testdata(test_pos, test_neg):
    neg_pairs = []
    for i in range(len(test_neg)):
        u, neg_list = test_pos['userID'].values[i], test_neg.iloc[i][1:]
        #if len(neg_list) != 999:
        #rem_idx = random.choice(range(1,len(neg_list)+1))
        #del neg_list[rem_idx]
        #print(len(neg_list))
        neg_pairs.extend([[u,b,0,0] for b in neg_list])
    neg_valid = pd.DataFrame(data=neg_pairs,columns = test_pos.columns)
    test_data = test_pos.append(neg_valid).reset_index(drop=True)
    return test_data
```

## Re-map

```python
print("before:", len(train))
data = train[train["rating"]>=5.0]
print("after:", len(data))
```

```python
user_map = dict(zip(data["userID"].unique(), range(data["userID"].nunique())))
item_map = dict(zip(data["bookID"].unique(), range(data["bookID"].nunique())))
data["userID"] = data["userID"].apply(lambda x: user_map[x])
data["bookID"] = data["bookID"].apply(lambda x: item_map[x])
assert data["userID"].nunique() == data["userID"].max() + 1
assert data["bookID"].nunique() == data["bookID"].max() + 1
```

```python
data.head()
```

```python
DEFAULT_USER_COL = 'userID'
DEFAULT_TIMESTAMP_COL = 'timestamp'
def leave_k_out_split(data, k=1, by=DEFAULT_USER_COL, chrono=True, margin=1):
    if chrono == False or len(data.columns.to_list()) <= 3:
        data = data.sample(frac=1)
        df_grouped = data.groupby(by)
    else:
        df_grouped = data.sort_values(DEFAULT_TIMESTAMP_COL).groupby(by)            
    leave_index = []
    for name, group in df_grouped:
        index = group.index
        if len(index) >= (k+margin):
            leave_index.extend(index[-k:])
    test = data.iloc[leave_index]
    train = data.iloc[list(set(data.index)-set(leave_index))] 
    print("leave_k_out split for %d/%d group(s)!" % (len(leave_index)//k, len(df_grouped)))
    return train.reset_index(drop=True), test.reset_index(drop=True)
```

```python
def get_traintest(data1):
    _, _, _, _, _, usernotread1 = get_embedding(data1)
    train1, test1 = leave_k_out_split(data1)
    neg_sample1 = 99
    neg_pairs1 = []
    for user in test1[DEFAULT_USER_COL].values:
        neg_pairs1.extend([[user,i,0,0] for i in random.sample(usernotread1[user], neg_sample1)])
    neg_valid1 = pd.DataFrame(data=neg_pairs1,columns = train1.columns)
    test1 = test1.append(neg_valid1).reset_index(drop=True)
    return train1, test1
```

```python
def get_embedding(data):
    users, books = data['userID'].values, data['bookID'].values 
    user_books, usernotread, usernotread_idx = defaultdict(set), defaultdict(set), defaultdict(set)
    all_users, all_books = list(set(users)), list(set(books))
    user2index, book2index = defaultdict(int), defaultdict(int)
    for i in range(len(users)):
        user_books[users[i]].add(books[i])
    
    for b_idx in range(len(all_books)):
        book2index[all_books[b_idx]] = b_idx

    all_books = set(all_books)
    # embedding
    for u_idx in range(len(all_users)):
        u = all_users[u_idx]
        user2index[u] = u_idx
        usernotread[u] = all_books-user_books[u]
        usernotread_idx[u_idx] = set([book2index[b] for b in usernotread[u]])
        usernotread[u] = list(usernotread[u])
         
    all_books = list(all_books)      
    return all_users, all_books, user_books, user2index, book2index, usernotread
```

```python
class BPR_MF:
    def __init__(self, p_lambda, lr, max_epoch, dimension, eva_k, batch_size):
        self.p_da = p_lambda
        self.lr = lr
        self.epo = max_epoch
        self.k = dimension
        self.eva_k = eva_k
        self.bs = batch_size

    def predict(self, u, b):
        return self.w[self.u2i[u]].dot(self.h[self.b2i[b]].T)
    
    def fit(self, train_data, Xvalid, yvalid, record_epoch):
        self.all_u, self.all_b, self.u_b, self.u2i, self.b2i, self.usernotread = get_embedding(train_data)
        self.w = np.random.normal(0,0.01,(len(self.all_u),self.k))
        self.h = np.random.normal(0,0.01,(len(self.all_b),self.k))
        ones = np.ones((self.k, 1))
        self.test_user_book_rank = test_u_b_rank(Xvalid, yvalid)
        self.hr, self.ndcg, self.auc = [], [], []
        self.record_w, self.record_h = [], []
        accu_w, accu_h = [0]*self.k, [0]*self.k
        
        for epoch in tqdm_notebook(range(self.epo)):
            user_list, item_pos_list, item_neg_list = self.shuffle(train_data) # shuffle 
            for i in range(len(user_list)): # length of data divides batch_size
                u_idx_batch, i_idx_batch, j_idx_batch = user_list[i], item_pos_list[i], item_neg_list[i]           
                x_uij_batch = (self.w[u_idx_batch] * (self.h[i_idx_batch]-self.h[j_idx_batch])).dot(ones)
                sigmoid_x_batch = 1/(1+np.exp(x_uij_batch))
                # update w and h
                w_diff = sigmoid_x_batch*(self.h[j_idx_batch]-self.h[i_idx_batch]) + self.p_da*self.w[u_idx_batch]
                i_diff = sigmoid_x_batch*(-self.w[u_idx_batch]) + self.p_da*self.h[i_idx_batch]
                j_diff = sigmoid_x_batch*self.w[u_idx_batch] + self.p_da*self.h[j_idx_batch]
                self.w[u_idx_batch] -= self.lr*w_diff 
                self.h[i_idx_batch] -= self.lr*i_diff
                self.h[j_idx_batch] -= self.lr*j_diff
            
            #if (epoch+1) % 10 == 0:
            ndcg, hr, auc = evaluation(self.test_user_book_rank, self.predict, self.eva_k)
            self.hr.append(hr)
            self.ndcg.append(ndcg)
            self.auc.append(auc)
            print(f'epoch: {epoch+1}, HR: {hr}, NDCG: {ndcg}, AUC: {auc}')
            if epoch == record_epoch:
                self.record_w = self.w.copy()
                self.record_h = self.h.copy()
       
    def shuffle(self, dataset):
        user_input = dataset['userID'].values
        item_input_pos = dataset['bookID'].values
        index = list(range(len(user_input)))
        np.random.shuffle(index)
        num_batch = len(user_input) // self.bs
        user_list, item_pos_list, item_neg_list = [], [], []
        for i in range(num_batch):
            begin = i * self.bs
            user_batch = [self.u2i[u] for u in user_input[index[begin:begin+self.bs]]]
            item_batch = [self.b2i[b] for b in item_input_pos[index[begin:begin+self.bs]]]
            item_neg_batch = [self.b2i[random.choice(self.usernotread[u])] for u in user_input[index[begin:begin+self.bs]]]
                
            user_list.append(user_batch)
            item_pos_list.append(item_batch)
            item_neg_list.append(item_neg_batch)
        return user_list, item_pos_list, item_neg_list


class BPR_GMF(BPR_MF):
    def __init__(self, p_lambda, lr, max_epoch, dimension, eva_k, batch_size, ratio=1e-2):
        super(BPR_GMF, self).__init__(p_lambda, lr, max_epoch, dimension, eva_k, batch_size)
        self.ratio = ratio

    def predict(self, u, b):
        return self.g.dot(self.w[self.u2i[u]] * (self.h[self.b2i[b]]))
    
    def fit(self, train_data, Xvalid, yvalid, record_epoch):
        self.all_u, self.all_b, self.u_b, self.u2i, self.b2i, self.usernotread = get_embedding(train_data)
        self.w = np.random.normal(0,0.01,(len(self.all_u),self.k))
        self.h = np.random.normal(0,0.01,(len(self.all_b),self.k))
        self.g = np.ones(self.k)
        self.test_user_book_rank = test_u_b_rank(Xvalid, yvalid)
        self.hr, self.ndcg, self.auc = [], [], []
        self.record_w, self.record_h = [], []
        
        for epoch in tqdm_notebook(range(self.epo)):
            user_list, item_pos_list, item_neg_list = self.shuffle(train_data) # shuffle
            for i in range(len(user_list)): # length of data divides batch_size
                u_idx_batch, i_idx_batch, j_idx_batch = user_list[i], item_pos_list[i], item_neg_list[i]           
                x_uij_batch = (self.w[u_idx_batch] * (self.h[i_idx_batch]-self.h[j_idx_batch])).dot(self.g)
                sigmoid_x_batch = 1/(1+np.exp(x_uij_batch))[:, None] # (64,)
                # update g, w and h
                diff_g = self.ratio*self.lr*(np.mean(sigmoid_x_batch*self.w[u_idx_batch]*(self.h[j_idx_batch]-self.h[i_idx_batch]), axis=0))
                diff_w = self.lr*(sigmoid_x_batch*self.g*(self.h[j_idx_batch]-self.h[i_idx_batch]) + self.p_da*self.w[u_idx_batch])
                diff_i = self.lr*(sigmoid_x_batch*self.g*(-self.w[u_idx_batch]) + self.p_da*self.h[i_idx_batch])
                diff_j = self.lr*(sigmoid_x_batch*self.g*self.w[u_idx_batch] + self.p_da*self.h[j_idx_batch])
                self.g -= diff_g
                self.w[u_idx_batch] -= diff_w
                self.h[i_idx_batch] -= diff_i
                self.h[j_idx_batch] -= diff_j
            
            ndcg, hr, auc = evaluation(self.test_user_book_rank, self.predict, self.eva_k)
            self.hr.append(hr)
            self.ndcg.append(ndcg)
            self.auc.append(auc)
            if epoch == record_epoch:
                self.record_w = self.w.copy()
                self.record_h = self.h.copy()
            print(f'epoch: {epoch+1}, HR: {hr}, NDCG: {ndcg}, AUC: {auc}')
```
